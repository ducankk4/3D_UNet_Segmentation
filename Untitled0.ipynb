#Preprocessing

```python
t1ce_list = sorted(glob.glob(F"{DATA_PATH}/*/*t1c.nii"))
t2_list = sorted(glob.glob(f"{DATA_PATH}/*/*t2w.nii"))
flair_list = sorted(glob.glob(f"{DATA_PATH}/*/*t2f.nii"))
mask_list = sorted(glob.glob(f"{DATA_PATH}/*/*seg.nii"))

print(f"len t1ce: {len(t1ce_list)}")
```

#Chọn lọc ảnh đủ tiêu chuẩn, tách ảnh, gộp 3 loại t1ce, t2,flair thành tensor

```python
scaler = MinMaxScaler()
def scale_img(path):
  img_data = nib.load(path).get_fdata()
  img_data_scaled = scaler.fit_transform(img_data.reshape(-1,img_data.shape[-1])).reshape(img_data.shape)
  return img_data_scaled
```

```python
tem_file = nib.load(t1ce_list[0]).get_fdata()
print(tem_file.shape)
print(tem_file.shape[-1])
```

```python

for idx in tqdm(range(len(t1ce_list)), unit= "file"):
  #tqdm (hien thi thanh tien trinh don vi tinh la file(unit))

  temp_mask = nib.load(mask_list[idx]).get_fdata()
  temp_mask_sliced = temp_mask[56:184, 56:184, 13:141]
  val, count = np.unique(temp_mask_sliced, return_counts= True)

  # loai bo cac anh co ti le nhan 0/sum > 0.99
  if(1- (count[0] / count.sum())) >0.01:
    # one hot mask_img
    temp_mask_scaled = F.one_hot(torch.tensor(temp_mask_sliced, dtype= torch.long), num_classes= 4)

    # MinMaxScale cac anh
    temp_t1ce_scaled = scale_img(t1ce_list[idx])
    temp_t2_scaled = scale_img(t2_list[idx])
    temp_flair_scaled = scale_img(flair_list[idx])

    # gop anh
    temp_combined_img = np.stack(
        [temp_flair_scaled, temp_t1ce_scaled, temp_t2_scaled], axis= 3
    )

    # cat anh theo thuc nghiem chua nhieu thong tin nhat
    temp_combined_img_sliced = temp_combined_img[56:184, 56:184, 13:141]

    os.makedirs(IMG_OUTPUT__PATH, exist_ok= True)
    os.makedirs(MASK_OUTPUT__PATH, exist_ok= True)

    np.save(
        f"{IMG_OUTPUT__PATH}/image_{str(idx)}.npy",
        temp_combined_img_sliced
    )
    np.save(
        f"{MASK_OUTPUT__PATH}/mask_{str(idx)}.npy",
        temp_mask_scaled
    )

  else:
    pass



```
